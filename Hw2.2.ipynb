Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

```python
programming_language = 'Python'
python_data = df[df['Мова програмування'] == programming_language]

print(python_data)
```

Визначте розмір таблиці python_data за допомогою методу shape

```python
python_data.shape
```

Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
python_data.groupby(['Посада'], as_index=False)
```

Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
salary = python_data.groupby(['Посада']).agg({'Зарплата на місяць': ['min', 'max']})
salary.columns = ['Min salary', 'Max salary']
salary
```

Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
salary['avg'] = salary.apply(lambda row: (row['Min salary'] + row['Max salary']) / 2, axis=1)
salary
```

Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
salary['avg'].describe()
```

Збережіть отриману таблицю в CSV файл

```python
salary.to_csv('salary.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Кількість чоловіків і жінок
gender_counts = df['Стать'].value_counts()

# Оформлення для кругової діаграми
colors = ['#66b3ff', '#ff9999']  # Кольори для чоловіків і жінок
explode = (0.1, 0)  # Висувні підсумки (розділення для "жінок")
labels = gender_counts.index.tolist()  # Назви категорій

# Створення кругової діаграми
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 12})
plt.title('Розподіл статі програмістів', fontsize=16)
plt.axis('equal')  # Забезпечити кругову форму
plt.tight_layout()  # Покращення розташування елементів
plt.show()
```

```python
avg_salary_by_position = python_data.groupby('Посада')['Зарплата на місяць'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
avg_salary_by_position.plot(kind='bar', color='skyblue')
plt.title('Середня зарплата за місяць за посадами', fontsize=16)
plt.xlabel('Посада', fontsize=12)
plt.ylabel('Середня зарплата на місяць', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
