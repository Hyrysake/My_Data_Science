```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
source = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C'
df = pd.read_html(source, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]
df
```

Вивести перші рядки таблиці за допомогою методу head

```python
df.head(5)
```

Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
print("\nКількість рядків та стовпців:", df.shape)
```

Замініть у таблиці значення "—" на значення NaN

```python
ukraine_birth_rate = df.iloc[-1, 1:].astype(float)
plt.figure(figsize=(10, 6))
ukraine_birth_rate.plot(kind='barh', color='skyblue')
plt.title('Народжуваність по Україні за роками')
plt.xlabel('Народжуваність')
plt.ylabel('Рік')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df.replace('—', np.nan, inplace=True)
df
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
df.dtypes
```

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
years_to_change = ["2014", "2019"]
df[years_to_change ] = df[years_to_change ].apply(pd.to_numeric, errors='coerce')
df.dtypes
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
share_of_gaps = (df.isnull().sum() / len(df)) * 100
share_of_gaps
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df =df.drop(df.index[-1])
df
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
df=df.fillna(df.mean(numeric_only=True)).round(1)
df
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
avg = df['2019'].mean()
df.loc[df['2019'] > avg]
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
max_born_2014 = df['2014'].max()
max_born_2014_region = df[df['2014']==max_born_2014]
result = max_born_2014_region['Регіон']
print(result.values[0])
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Регіон'], df['2019'])
plt.xticks(rotation=90)
plt.xlabel('Регіон')
plt.ylabel('Рівень народжуваності у 2019 році')
plt.title('Рівень народжуваності по регіонах у 2019 році')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df.columns[1:], df[df['Регіон'] == 'Волинська'].iloc[0, 1:], marker='o', linestyle='-', color='blue')
plt.title('Народжуваність у Волинській області за роками')
plt.xlabel('Рік')
plt.ylabel('Народжуваність')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df.columns[1:], df[df['Регіон'] == 'Львівська'].iloc[0, 1:], color='green')
plt.title('Народжуваність у Львові за роками')
plt.xlabel('Рік')
plt.ylabel('Народжуваність')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
